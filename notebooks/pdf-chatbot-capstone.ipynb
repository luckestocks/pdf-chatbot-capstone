```python
# Step 2 — Connecting to Groq and talk to Llama-3
from google.colab import userdata
from groq import Groq

# This reads your API key safely from Colab Secrets
api_key = userdata.get('GROQ_API_KEY')

# Creating a connection to Groq
client = Groq(api_key=api_key)

# Sending a test message to Llama-3
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Say hello and tell me you are Llama-3 running on Groq!"}
    ]
)

# Print the response
print(response.choices[0].message.content)
```
